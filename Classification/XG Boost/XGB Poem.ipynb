```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('all.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.columns = ['author','content','PoemName','age','type']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
data.author = enc.fit_transform(data.author)
data.content = enc.fit_transform(data.content)
data.age = enc.fit_transform(data.age)
data.type = enc.fit_transform(data.type)
data.PoemName = enc.fit_transform(data.PoemName)
```

```python
data.isnull().sum()
```

```python
M = data['PoemName'].mean()
M
```

```python
data['PoemName'].fillna(value=M,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
y = data.PoemName
X = data.iloc[:,[0,1,3,4]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=5)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,predict)
```

```python
confusion_matrix(y_test,predict)
```

```python
print(classification_report(y_test,predict))
```
